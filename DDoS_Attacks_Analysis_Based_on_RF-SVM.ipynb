```
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')

dataset = pd.read_csv('/content/drive/My Drive/DDoS/kdd.csv')
dataset.drop_duplicates(subset=None,keep='first',inplace=True)
x=dataset.drop(columns=list(dataset.columns)[-1])
y=dataset[list(dataset.columns)[-1]]
```

```
x
```

```
le= preprocessing.LabelEncoder()
x['protocol_type']=le.fit_transform(x['protocol_type'])
x['service']=le.fit_transform(x['service'])
x['flag']=le.fit_transform(x['flag'])
y=le.fit_transform(y)


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1000)
```

```
nor=preprocessing.Normalizer()
x_train=nor.fit_transform(x_train)
x_test=nor.fit_transform(x_test)
```

```
x
```

```
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf.fit(x_train,y_train)
pred=clf.predict(x_test)
```

```
tr=dataset.drop(columns=list(dataset.columns)[-1])
feat_array=sorted(np.c_[range(len(tr.columns)),tr.columns,clf.feature_importances_],reverse=True,key=lambda x:x[2])
for i in feat_array:
  print(i)
```

```
alpha=0.01
A_index=[]
B_index=[]
max_feature=0.0
for feature in feat_array:
    if feature[2]>alpha:
        A_index.append(feature[0])
        if max_feature<feature[2]:
               max_feature=feature[2]
    else:
        B_index.append(feature[0])
```

```
A_index
```

```
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import precision_score,accuracy_score,recall_score
sum1=0.00
k=0
for xi in B_index:
        k+=1
        clfA = SVC(kernel='linear',C=1.0,gamma=0.2,cache_size=7000)
        clfB = SVC(kernel='linear',C=1.0,gamma=0.2,cache_size=7000)
        print("SVC initialization done");
        clfA.fit(x_train[:,A_index],y_train)
        print(str(k)," SVM on A is done........")
        clfB.fit(x_train[:,A_index+[xi]],y_train)
        print(str(k)," SVM on B is done........")
        predictA=clfA.predict(x_test[:,A_index])
        predictB=clfB.predict(x_test[:,A_index+[xi]])
        first=precision_score(y_test,predictA,average="micro")
        second=precision_score(y_test,predictB,average="micro")
        resu=second-first
        print(resu)
        if(resu>0.000000001):
          A_index.append(xi)
```

```
clfran = SVC(kernel='linear')
clfran.fit(x_train[:,A_index],y_train)
predran=clfran.predict(x_test[:,A_index])
```

```
from sklearn.model_selection import cross_val_score
acc1 =accuracy_score(y_test,predran) 
acc1=cross_val_score(clfran,x_test[:,A_index],y_test,scoring='accuracy',cv=10)
acc1=np.mean(acc1)
```

```
prec1 =precision_score(y_test,predran,average='micro') 
prec1=cross_val_score(clfran,x_test[:,A_index],y_test,scoring='precision_weighted',cv=10)
prec1=np.mean(prec1)
print(prec1) #Precision of RF-SVM
reca1 = recall_score(y_test,predran,average='micro')
reca1=cross_val_score(clfran,x_test[:,A_index],y_test,scoring='recall_weighted',cv=10)
reca1=np.mean(reca1) #Recall of RF-SVM
```

```
cl= SVC(kernel='linear')
cl.fit(x_train,y_train)
predictr=cl.predict(x_test)
acc2=cross_val_score(cl,x_test,y_test,scoring='accuracy',cv=10)
acc2=np.mean(acc2)
prec2=cross_val_score(cl,x_test,y_test,scoring='precision_weighted',cv=10)
prec2=np.mean(prec2)
reca2=cross_val_score(cl,x_test,y_test,scoring='recall_micro',cv=10)
reca2=np.mean(reca2)
print(acc2)
print(prec2)
print(reca2)
```

```
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
predictions=logmodel.predict(x_test)
acc3=accuracy_score(y_test,predictions)
print(acc3) #Accuracy of Logistic regression
```

```
prec3=precision_score(y_test,predictions,average='micro')
print(prec3) #Precision of Logistic regression
reca3=recall_score(y_test,predictions,average='micro')
print(reca3) #Recall of Logistic Regression
```

```
from sklearn.ensemble import AdaBoostClassifier
adamodel= AdaBoostClassifier(n_estimators=20,learning_rate=1)
adamodel.fit(x_train,y_train)
y_pred=adamodel.predict(x_test)
```

```
acc4=accuracy_score(y_test,y_pred)
print(acc4) #Accuracy of Adaboost
prec4 = precision_score(y_test,y_pred,average='micro')
print(prec4) #Precision of Adaboost
reca4=recall_score(y_test,y_pred,average='micro') #Recall of SVM
print(reca4) #Recall of Adaboost
```
